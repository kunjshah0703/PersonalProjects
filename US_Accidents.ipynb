```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualising 20 most cities based on number of accidents
x_cities = accidents_copy['City'].value_counts()[:20].index
y_acc_values = accidents_copy['City'].value_counts()[:20].values
plt.figure(figsize = (12,8))
plt.bar(x = x_cities, height = y_acc_values, color = ['Red', 'Orange', 'Pink', 'Yellow', 'Green', 'Blue', 'Purple', 'Black', 'Maroon', 'Grey'], width = 0.9)
plt.xticks(rotation = 45)
plt.title('Number of accidents based on city')
plt.xlabel('Cities')
plt.ylabel('Number of Accidents')
for i in range(len(x_cities)):
    plt.text(i, y_acc_values[i], y_acc_values[i], ha = 'center', va = 'bottom', fontsize = 8)
plt.show()
```
