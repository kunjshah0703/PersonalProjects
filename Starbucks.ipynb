```python
x_prep = df_drink_menu['Beverage_prep'].value_counts().index
y_prep_value = df_drink_menu['Beverage_prep'].value_counts().values
```

```python
plt.figure(figsize=(12,8))
colors = ["red" if i > 20 else "green" for i in y_prep_value]
plt.bar(x = x_prep, height=y_prep_value, color = colors)
plt.title('Beverage prep count')
plt.xlabel('Types of Prep Bevergaes')
plt.ylabel('Count of type of prep Beverages')
plt.xticks(rotation = 60)
for i in range(len(x_prep)):
    plt.text(i, y_prep_value[i], y_prep_value[i], ha = 'center', va = 'bottom', fontsize = 10)
plt.show()
```
