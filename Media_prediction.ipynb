```python
x_foods = media_prediction_copy['food_category'].value_counts().sort_values(ascending=False)[:20].index
y_food_count = media_prediction_copy['food_category'].value_counts().sort_values(ascending=False)[:20].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Analyzing top 20 food categories
plt.figure(figsize=(12,8))
plt.bar(x = x_foods, height = y_food_count)
plt.xticks(rotation = 90)
plt.xlabel("Food Category", fontweight = 'bold')
plt.ylabel("Count of Food Category", fontweight = 'bold')
plt.title("Count of Food Category", fontweight = 'bold')
for i in range(len(x_foods)):
    plt.text(i, y_food_count[i], y_food_count[i], ha = 'center', va = 'bottom')
plt.show()
```
