```python
x_ocean_type = df_copy['ocean_proximity'].value_counts().index
```

```python
y_ocean_value = df_copy['ocean_proximity'].value_counts().values
```

```python
plt.figure(figsize=(10,5))
colors = ["red" if i > 6000 else "green" for i in y_ocean_value]
plt.bar(x = x_ocean_type, height = y_ocean_value, width = 0.5, color = colors)
plt.title("Number of houses in a particular location")
plt.xlabel('Location Types')
plt.ylabel('Count of properties')
for i in range(len(x_ocean_type)) : 
    plt.text(i , y_ocean_value[i], y_ocean_value[i], ha = 'center', va = 'bottom', fontsize = 10)
plt.show()
```
