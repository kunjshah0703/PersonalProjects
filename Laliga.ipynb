```python
player = laliga_copy.groupby('Name')['Yellow Cards'].sum().sort_values(ascending=False)[:10].index
yellow_cards_count = laliga_copy.groupby('Name')['Yellow Cards'].sum().sort_values(ascending=False)[:10].values

plt.figure(figsize=(13,5))
colors = ["red" if i >= 14 else "green" for i in yellow_cards_count]
plt.bar(x = player, height = yellow_cards_count, color = colors)
plt.xticks(rotation = 75)
plt.title("Top 10 players with most numbe of yellow cards in the season")
plt.xlabel('Player Name')
plt.ylabel('Count of Yellow Cards')
for i in range(len(player)):
    plt.text(i, yellow_cards_count[i], yellow_cards_count[i], ha = 'center', va = 'bottom')
plt.show()

plt.pie(yellow_cards_count, labels=player, autopct='%1.2f%%')
plt.title('Perecntage contribution of Top 10 Players with most yellow cards')
plt.show()
```

```python
x_team_ta = laliga_copy.groupby('Team')['Tackles'].sum().sort_values(ascending=False).index
y_team_tackles = laliga_copy.groupby('Team')['Tackles'].sum().sort_values(ascending=False).values
```

```python
plt.figure(figsize=(12,6))
plt.bar(x = x_team_ta, height = y_team_tackles)
plt.title('Number of tackles by a team in a season')
plt.xlabel('Teams')
plt.ylabel('Number of Tackles')
plt.xticks(rotation = 70)
for i in range(len(x_team_ta)):
    plt.text(i, y_team_tackles[i], y_team_tackles[i], ha = 'center', va = 'bottom')
plt.show()
```

```python
y_unsuccess_tackles = laliga_copy.groupby('Team')['Unssuccessful tackles'].sum().sort_values(ascending=False).values
```

```python
y_unsuccess_tackles
```

```python
y_success_tackles = laliga_copy.groupby('Team')['Successful tackles'].sum().sort_values(ascending=False).values
```

```python
y_success_tackles
```

```python
plt.figure(figsize=(12,6))
plt.bar(x_team_ta, y_success_tackles, bottom = y_unsuccess_tackles, color = 'green', label = 'Successful')
for i in range(len(x_team_ta)):
    plt.text(i, y_success_tackles[i], y_success_tackles[i], ha = 'center', va = 'top')
plt.bar(x_team_ta, y_unsuccess_tackles, color = 'red', label = 'Unsucessful')
for j in range(len(x_team_ta)):
    plt.text(j, y_unsuccess_tackles[j], y_unsuccess_tackles[j], ha = 'center', va = 'bottom')
plt.xticks(rotation = 60)
plt.xlabel('Teams')
plt.ylabel('Number of Successful Tackles vs Unsuccessful Tackles')
plt.title('Number Successful and Unsuccessful tackles by a Team in a Season')
plt.legend()
plt.show()
```
